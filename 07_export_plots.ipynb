# 7. Data Export & Archiving

**Summary**: Zips all generated plots and data into archives for sharing or download.

**Key Steps**:
1. Create `plots_and_data.zip`.

**Inputs**:
- `plots/` directory
- `data/` directory

**Outputs**:
- `plots_and_data.zip`

**Runtime**: 
- creating takes about 21 seconds, 
- downloading (over ssh) takes much longer: 4+ minutes

---

```python
# Create zip archive of the plots and data folders
# creating takes about 21 seconds
# downloading much longer: 4+ minutes
import subprocess
import os

zip_filename = 'plots_and_data.zip'

if os.path.exists(zip_filename):
    os.remove(zip_filename)

cmd = ['zip', '-r', zip_filename, 'plots', 'data', '-x', '*.pkl']

print(f"Running: {' '.join(cmd)}")
result = subprocess.run(cmd, capture_output=True, text=True)

if result.returncode == 0:
    print(f"Successfully created {zip_filename}")
    # Print some info about content (optional)
    # print(result.stdout)
else:
    print("Error creating zip:")
    print(result.stderr)
```
